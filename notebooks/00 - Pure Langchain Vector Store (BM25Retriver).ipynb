```python
# !pip install -q langchain langchain-community langchain-openai rank-bm25
from dotenv import load_dotenv
import os
from langchain_core.documents import Document
from langchain_community.retrievers import BM25Retriever
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

# Set your API key however you prefer
load_dotenv()
# os.environ["OPENAI_API_KEY"] = "your-key"

# 1) Your small document set
docs = [
    Document(page_content="LangChain is a framework for building LLM-powered applications."),
    Document(page_content="RAG stands for Retrieval-Augmented Generation."),
    Document(page_content="BM25 is a sparse retrieval algorithm based on keyword matching."),
    Document(page_content="A vector database is useful for semantic search, but not required for a simple RAG demo."),
    Document(page_content="Retrievers fetch relevant context before the model answers a question."),
]

# 2) Build a retriever without any vector DB
retriever = BM25Retriever.from_documents(docs)
retriever.k = 2  # number of chunks/documents to retrieve

# 3) Create the prompt
prompt = ChatPromptTemplate.from_template(
    """Answer the question using only the context below.

Context:
{context}

Question:
{question}

If the answer is not in the context, say you don't know."""
)

# 4) LLM
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# 5) Ask a question
question = "Do I need a vector database for a simple RAG demo?"

# Retrieve documents
retrieved_docs = retriever.invoke(question)
context = "\n\n".join(doc.page_content for doc in retrieved_docs)

# Generate answer
messages = prompt.invoke({"context": context, "question": question})
response = llm.invoke(messages)

print("Retrieved context:\n")
for i, doc in enumerate(retrieved_docs, 1):
    print(f"{i}. {doc.page_content}")

print("\nAnswer:\n")
print(response.content)
```
